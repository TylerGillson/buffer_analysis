```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
varied_buff_size = pd.read_csv("/Users/tylergillson/Dropbox/UofC/F2018/CPSC 441/Assignments/A3/results.csv")
varied_tr_out = pd.read_csv("/Users/tylergillson/Dropbox/UofC/F2018/CPSC 441/Assignments/A3/tr_rate.csv")
```

```python
varied_buff_size['BuffSize'] = [i for i in range(1000)]
varied_tr_out['TransmissionRate'] = [i for i in range(101)]
```

```python
varied_buff_size.head(2)
```

```python
varied_tr_out.head(2)
```

```python
dims = (16, 8)
fig, ax = plt.subplots(figsize=dims)

g = sns.lineplot(x='BuffSize', y='PktLossPct', data=varied_buff_size, color="green")
g.set_xticks([i for i in range(1001) if i % 50 == 0])
ax2 = plt.twinx()
g2 = sns.lineplot(x='BuffSize', y='AvgQueuingDelay', data=varied_buff_size, ax=ax2, color="red")

g.axes.set_title("Buffer Size vs. QoS",fontsize=40)
g.set_xlabel("Buffer Size",fontsize=20)
g.set_ylabel("Packet Loss %",fontsize=20)
g2.set_ylabel("Avg. Queuing Delay (sec)",fontsize=20)
```

```python
dims = (16, 8)
fig, ax = plt.subplots(figsize=dims)

g = sns.lineplot(x='TransmissionRate', y='PktLossPct', data=varied_tr_out, color="green")
g.set_xticks([i for i in range(101) if i % 5 == 0])
g.set_yticks([i for i in range(101) if i % 5 == 0])
ax2 = plt.twinx()
g2 = sns.lineplot(x='TransmissionRate', y='AvgQueuingDelay', data=varied_tr_out, ax=ax2, color="red")

g.axes.set_title("Transmission Rate vs. QoS",fontsize=40)
g.set_xlabel("Transmission Rate (Mbps)",fontsize=20)
g.set_ylabel("Packet Loss %",fontsize=20)
g2.set_ylabel("Avg. Queuing Delay (sec)",fontsize=20)
```
